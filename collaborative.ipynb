```python
data=data.dropna(how="any")
data['Description']=data['Description'].replace({"'ll":' '},regex=True)
```

```python
data['CourseId']
```

```python
data['CourseId']=data['CourseId'].replace({"-":' '},regex=True)
data['CourseId']
```

```python
newdata=data.CourseId.str.cat(' '+data.CourseTitle.str.cat(' '+data.Description))
```

```python
newdata=newdata.replace({"[^A-Za-z0-9]+": ""},regex=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectoriser=TfidfVectorizer(stop_words='english')
```

```python
X=vectoriser.fit_transform(newdata)
print(X)
```

```python
from sklearn.cluster import KMeans
true_k=8
model=KMeans(n_clusters=true_k,init='k-means++',max_iter=500,n_init=15).fit(X)
```

```python
print("top terms per cluster")
doc_centroid=model.cluster_centers_.argsort()[:,::-1]
terms=vectoriser.get_feature_names()
for i in range(true_k):
    print("cluster %d " %i )
    for j in doc_centroid[i,:15]:
        print(' %s' % terms[j]),
    
```
